```python
"""
GRU Autoencoder + Federated Learning (simulated FedAvg) + RF detector.
- Uses numeric columns only (drops non-numeric).
- If no anomaly label present, creates synthetic anomalies from a numeric proxy.
- Creates sequence windows of length 1 (easy to change to seq_len > 1).
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, GRU, RepeatVector, TimeDistributed, Dense

# ---------------- User settings ----------------
CSV_PATH = "/content/merged_user_features.csv"
POSSIBLE_LABELS = ["label", "is_anomaly", "is_anomalous", "anomaly", "target", "y"]
ANOMALY_FRAC = 0.10        # synthetic anomaly fraction if no label
N_CLIENTS = 5
ROUNDS = 3                 # federated rounds (increase for better results)
LOCAL_EPOCHS = 2           # local epochs per client
BATCH_SIZE = 64
LR = 1e-3
RANDOM_SEED = 42
TEST_SIZE = 0.2
SEQ_LEN = 1                # sequence length (1 = each row is a single timestep)
DEVICE = "/gpu:0" if tf.config.list_physical_devices("GPU") else "/cpu:0"
# ------------------------------------------------

np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

# ---------------- Load data ----------------
if not os.path.exists(CSV_PATH):
    raise FileNotFoundError(f"{CSV_PATH} not found.")

df = pd.read_csv(CSV_PATH)
print("Loaded data:", df.shape)
print("Columns:", list(df.columns))

# ---------------- Pick label (if any) OR build proxy ----------------
label_col = None
for nm in POSSIBLE_LABELS:
    if nm in df.columns:
        label_col = nm
        break

# Work on numeric columns only for model input
numeric_df = df.select_dtypes(include=[np.number]).copy()
if numeric_df.shape[1] == 0:
    raise ValueError("No numeric columns found. Convert/encode categorical columns first.")

if label_col:
    print(f"Using provided label column: {label_col}")
    y_raw = df[label_col].values
    # ensure label column not used as input
    if label_col in numeric_df.columns:
        numeric_df = numeric_df.drop(columns=[label_col])
    has_label = True
else:
    # build numeric proxy (row-sum) to derive anomalies
    y_proxy = numeric_df.sum(axis=1).values
    y_raw = y_proxy
    has_label = False
    print("No label column found — using numeric proxy (row-sum) to synthesize anomalies if needed.")

# ---------------- Build binary anomaly labels ----------------
if has_label:
    # if binary already
    uniq = np.unique(y_raw[~pd.isna(y_raw)])
    if set(uniq) <= {0, 1}:
        y_binary = np.nan_to_num(y_raw).astype(int)
    else:
        # convert top ANOMALY_FRAC of y_raw to anomalies
        thr = np.percentile(y_raw[~pd.isna(y_raw)], 100 * (1 - ANOMALY_FRAC))
        y_binary = (y_raw >= thr).astype(int)
else:
    thr = np.percentile(y_raw, 100 * (1 - ANOMALY_FRAC))
    y_binary = (y_raw >= thr).astype(int)

print("Anomaly class counts (0=normal,1=anomaly):", np.bincount(y_binary))

# ---------------- Preprocess inputs ----------------
X_df = numeric_df.fillna(0.0)
feature_names = X_df.columns.tolist()
X = X_df.values.astype(np.float32)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# reshape to sequences: (N, seq_len, n_features)
N, n_features = X_scaled.shape
X_seq = X_scaled.reshape((N, SEQ_LEN, n_features))   # seq_len default 1

# ---------------- Train/test split ----------------
stratify_labels = y_binary if len(np.unique(y_binary)) > 1 else None
train_idx, test_idx = train_test_split(np.arange(N), test_size=TEST_SIZE, random_state=RANDOM_SEED, stratify=stratify_labels)
X_train_seq = X_seq[train_idx]
X_test_seq  = X_seq[test_idx]
y_train = y_binary[train_idx]
y_test  = y_binary[test_idx]

print("Train/test sizes:", X_train_seq.shape[0], X_test_seq.shape[0])

# ---------------- Autoencoder builder ----------------
def build_gru_autoencoder(seq_len, n_features, enc_units=64, latent_dim=32):
    inp = Input(shape=(seq_len, n_features))
    encoded = GRU(enc_units, activation="tanh")(inp)           # (B, enc_units)
    latent = Dense(latent_dim, activation="relu")(encoded)    # (B, latent_dim)
    dec_in = Dense(enc_units, activation="relu")(latent)
    dec_in_rep = RepeatVector(seq_len)(dec_in)                # (B, seq_len, enc_units)
    dec_out = GRU(enc_units, activation="tanh", return_sequences=True)(dec_in_rep)
    out = TimeDistributed(Dense(n_features))(dec_out)
    model = Model(inputs=inp, outputs=out)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=LR), loss="mse")
    return model

# init global model
global_model = build_gru_autoencoder(SEQ_LEN, n_features, enc_units=64, latent_dim=32)

# ---------------- Federated training (simulated) ----------------
# create client splits (by indices)
indices = np.arange(X_train_seq.shape[0])
np.random.shuffle(indices)
client_splits = np.array_split(indices, N_CLIENTS)
print("Client sample counts:", [len(s) for s in client_splits])

# helper to aggregate weights (FedAvg weighted by number of samples)
def fedavg_aggregate(list_of_weight_lists, sizes):
    total = float(np.sum(sizes))
    # initialize with zeros arrays same shape as first client's weights
    avg = []
    for layer_idx in range(len(list_of_weight_lists[0])):
        # accumulate weighted arrays
        accum = np.zeros_like(list_of_weight_lists[0][layer_idx], dtype=np.float64)
        for client_idx, wlist in enumerate(list_of_weight_lists):
            accum += (wlist[layer_idx].astype(np.float64) * (sizes[client_idx] / total))
        avg.append(accum.astype(np.float32))
    return avg

# Federated rounds
global_weights = global_model.get_weights()
for rnd in range(1, ROUNDS + 1):
    client_weights = []
    client_sizes = []
    print(f"\n----- Fed Round {rnd} -----")
    for c_idx, idxs in enumerate(client_splits):
        if len(idxs) == 0:
            continue
        # build local model and set global weights
        local_model = build_gru_autoencoder(SEQ_LEN, n_features, enc_units=64, latent_dim=32)
        local_model.set_weights(global_weights)
        # local training on client's data
        local_x = X_train_seq[idxs]
        local_model.fit(local_x, local_x, epochs=LOCAL_EPOCHS, batch_size=BATCH_SIZE, verbose=0)
        w = local_model.get_weights()
        client_weights.append(w)
        client_sizes.append(len(idxs))
        print(f" Client {c_idx}: trained on {len(idxs)} samples")
    if len(client_weights) == 0:
        print("No client weights collected this round; skipping aggregation.")
        continue
    # aggregate
    global_weights = fedavg_aggregate(client_weights, client_sizes)
    global_model.set_weights(global_weights)
    # quick eval on small subset to monitor progress
    quick_n = min(256, X_test_seq.shape[0])
    pred_q = global_model.predict(X_test_seq[:quick_n])
    mse_q = np.mean((pred_q - X_test_seq[:quick_n])**2)
    print(f" Quick test MSE (subset): {mse_q:.6f}")

# save global model (optional)
global_model.save("gru_ae_fed_global.keras", include_optimizer=False)
print("\nSaved global model to gru_ae_fed_global.keras")

# ---------------- Reconstruction errors ----------------
recon_test = global_model.predict(X_test_seq, batch_size=256)
recon_train = global_model.predict(X_train_seq, batch_size=256)

# per-sample MSE
per_sample_mse_test = np.mean((X_test_seq - recon_test)**2, axis=(1,2))
per_sample_mse_train = np.mean((X_train_seq - recon_train)**2, axis=(1,2))

# per-feature absolute reconstruction error (averaged over timesteps -> seq_len=1)
per_feature_err_tr = np.mean(np.abs(X_train_seq - recon_train), axis=1)   # shape (N_train, n_features)
per_feature_err_te = np.mean(np.abs(X_test_seq - recon_test), axis=1)     # shape (N_test, n_features)

# ---------------- Simple threshold detector ----------------
k = 3.0
thr = per_sample_mse_train.mean() + k * per_sample_mse_train.std()
y_pred_thr = (per_sample_mse_test >= thr).astype(int)
print("\nThreshold detector results:")
print(" Threshold:", thr)
print(" Accuracy:", accuracy_score(y_test, y_pred_thr))
print(" Confusion matrix:\n", confusion_matrix(y_test, y_pred_thr))

# ---------------- Train RandomForest on per-feature errors ----------------
# check we have both classes in training
if len(np.unique(y_train)) < 2:
    print("\nWarning: training labels have only one class. Skipping RF classifier & SHAP.")
else:
    rf = RandomForestClassifier(n_estimators=200, random_state=RANDOM_SEED)
    rf.fit(per_feature_err_tr, y_train)
    y_pred_rf = rf.predict(per_feature_err_te)
    acc_rf = accuracy_score(y_test, y_pred_rf)
    cm_rf = confusion_matrix(y_test, y_pred_rf)
    print("\nRandomForest detector results:")
    print(" Accuracy:", acc_rf)
    print(" Confusion matrix:\n", cm_rf)
    print("\nClassification report:\n", classification_report(y_test, y_pred_rf, digits=4))

    # plot confusion matrix
    plt.figure(figsize=(5,4))
    sns.heatmap(cm_rf, annot=True, fmt="d", cmap="Blues")
    plt.xlabel("Predicted"); plt.ylabel("True"); plt.title("Confusion Matrix - RF on reconstruction errors")
    plt.show()

print("\nAll done.")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# ---------------- Load dataset ----------------
df = pd.read_csv("merged_user_features.csv")

# Drop non-numeric columns
X = df.drop(columns=["user", "date"])
X = X.fillna(0)  # handle missing values

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)

# Reshape into [samples, timesteps, features]
# We'll treat each row as a "sequence" of length 1 (can adapt if real sequences exist)
X_train_seq = np.expand_dims(X_train, axis=1)
X_test_seq = np.expand_dims(X_test, axis=1)

# Convert to PyTorch tensors
train_dataset = TensorDataset(torch.tensor(X_train_seq, dtype=torch.float32))
test_dataset = TensorDataset(torch.tensor(X_test_seq, dtype=torch.float32))
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# ---------------- GRU Autoencoder ----------------
class GRUAutoencoder(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, latent_dim=32, num_layers=1):
        super(GRUAutoencoder, self).__init__()
        self.encoder = nn.GRU(input_dim, hidden_dim, num_layers, batch_first=True)
        self.latent = nn.Linear(hidden_dim, latent_dim)
        self.decoder_gru = nn.GRU(latent_dim, hidden_dim, num_layers, batch_first=True)
        self.output_layer = nn.Linear(hidden_dim, input_dim)

    def forward(self, x):
        # Encoder
        _, h = self.encoder(x)  # take last hidden state
        z = self.latent(h[-1]).unsqueeze(1)  # latent vector

        # Decoder
        dec_out, _ = self.decoder_gru(z)
        out = self.output_layer(dec_out)
        return out

# ---------------- Training ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = GRUAutoencoder(input_dim=X_train.shape[1]).to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

n_epochs = 20
train_losses = []

for epoch in range(n_epochs):
    model.train()
    epoch_loss = 0
    for batch in train_loader:
        batch = batch[0].to(device)
        optimizer.zero_grad()
        output = model(batch)
        loss = criterion(output, batch)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    epoch_loss /= len(train_loader)
    train_losses.append(epoch_loss)
    print(f"Epoch {epoch+1}/{n_epochs}, Loss: {epoch_loss:.6f}")

# ---------------- Evaluation ----------------
model.eval()
recon_errors = []
with torch.no_grad():
    for batch in test_loader:
        batch = batch[0].to(device)
        output = model(batch)
        loss = criterion(output, batch)
        recon_errors.append(loss.item())

print("Mean Reconstruction Error:", np.mean(recon_errors))

# ---------------- Loss Curve ----------------
plt.plot(train_losses)
plt.xlabel("Epoch")
plt.ylabel("Reconstruction Loss")
plt.title("GRU Autoencoder Training Loss")
plt.show()
```

## Implement federated learning gru autoencoder

### Subtask:
Adapt the existing GRU autoencoder code to train in a federated learning setting using the preprocessed data.


**Reasoning**:
Implement the GRU Autoencoder model in PyTorch, set up the federated training loop, perform simulated federated training rounds, aggregate weights using FedAvg, update the global model, and save the global model after training.



```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import copy

# ---------------- GRU Autoencoder ----------------
class GRUAutoencoder(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, latent_dim=32, num_layers=1):
        super(GRUAutoencoder, self).__init__()
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.latent_dim = latent_dim
        self.num_layers = num_layers

        self.encoder = nn.GRU(input_dim, hidden_dim, num_layers, batch_first=True)
        self.enc_fc = nn.Linear(hidden_dim, latent_dim)

        # Decoder takes latent vector and repeats it for seq_len (which is 1 here)
        # to match the expected input shape for the GRU
        self.dec_fc = nn.Linear(latent_dim, hidden_dim)
        self.decoder = nn.GRU(hidden_dim, hidden_dim, num_layers, batch_first=True)
        self.output_layer = nn.Linear(hidden_dim, input_dim)

    def forward(self, x):
        # Encoder
        # x shape: (batch_size, seq_len, input_dim)
        _, h = self.encoder(x) # h shape: (num_layers, batch_size, hidden_dim)

        # Take the last hidden state from the last layer
        h_last = h[-1] # h_last shape: (batch_size, hidden_dim)

        # Latent representation
        z = self.enc_fc(h_last) # z shape: (batch_size, latent_dim)

        # Decoder input: repeat latent vector seq_len times (seq_len=1)
        dec_input = self.dec_fc(z).unsqueeze(1) # dec_input shape: (batch_size, 1, hidden_dim)

        # Decoder
        dec_out, _ = self.decoder(dec_input) # dec_out shape: (batch_size, 1, hidden_dim)

        # Output layer
        output = self.output_layer(dec_out.squeeze(1)).unsqueeze(1) # output shape: (batch_size, 1, input_dim)

        return output

# ---------------- Federated Learning Setup ----------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

input_dim = X_train_seq.shape[-1]
global_model = GRUAutoencoder(input_dim=input_dim).to(device)
criterion = nn.MSELoss()

# Simulate client data splits (based on the existing train_test_split indices)
n_clients = 5
# We need to split the *training* data indices for clients
indices = np.arange(X_train_seq.shape[0])
np.random.shuffle(indices) # Shuffle indices for randomness in client data
client_splits = np.array_split(indices, n_clients)

# ---------------- Federated Training (Simulated FedAvg) ----------------
n_rounds = 10 # Number of communication rounds
local_epochs = 3 # Epochs each client trains locally
lr = 0.001 # Learning rate

global_weights = global_model.state_dict()

for rnd in range(n_rounds):
    print(f"\n----- Fed Round {rnd + 1} -----")
    client_weights = []
    client_sizes = []

    for c_idx, idxs in enumerate(client_splits):
        if len(idxs) < 1: # Skip clients with no data
            continue

        print(f" Training Client {c_idx + 1}/{n_clients}...")

        # Create local model and load global weights
        local_model = GRUAutoencoder(input_dim=input_dim).to(device)
        local_model.load_state_dict(global_weights)

        # Create DataLoader for client's local data
        local_X = X_train_tensor[idxs]
        local_dataset = TensorDataset(local_X)
        local_loader = DataLoader(local_dataset, batch_size=BATCH_SIZE, shuffle=True)

        # Local training
        local_optimizer = torch.optim.Adam(local_model.parameters(), lr=lr)
        local_model.train()
        for ep in range(local_epochs):
            for batch in local_loader:
                batch_data = batch[0].to(device)
                local_optimizer.zero_grad()
                output = local_model(batch_data)
                loss = criterion(output, batch_data)
                loss.backward()
                local_optimizer.step()
            # print(f"  Client {c_idx+1}, Epoch {ep+1}/{local_epochs}, Loss: {loss.item():.6f}") # Optional: print local loss

        # Collect local model weights and data size
        client_weights.append(copy.deepcopy(local_model.state_dict()))
        client_sizes.append(len(idxs))
        del local_model # Free up memory

    if len(client_weights) == 0:
        print("No client weights collected this round; skipping aggregation.")
        continue

    # Aggregate weights (FedAvg)
    print(" Aggregating client weights...")
    aggregated_weights = {}
    for key in global_weights.keys():
        aggregated_weights[key] = torch.stack([client_weights[i][key].float() * client_sizes[i] for i in range(len(client_weights))], 0).sum(0) / sum(client_sizes)

    # Update global model weights
    global_model.load_state_dict(aggregated_weights)

    # Optional: Evaluate global model on a small subset of test data
    global_model.eval()
    with torch.no_grad():
        quick_n = min(256, X_test_tensor.shape[0])
        quick_test_data = X_test_tensor[:quick_n].to(device)
        quick_recon = global_model(quick_test_data)
        quick_mse = criterion(quick_recon, quick_test_data).item()
        print(f" Quick test MSE (subset): {quick_mse:.6f}")


# Save the final global model
torch.save(global_model.state_dict(), "gru_ae_fed_global_torch.pth")
print("\nSaved global model to gru_ae_fed_global_torch.pth")
```

## Train and evaluate autoencoder

### Subtask:
Train the federated GRU autoencoder and evaluate its performance based on reconstruction error.


**Reasoning**:
Load the saved global model state dictionary and calculate the reconstruction error on the test set to evaluate the model's performance.



```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
import numpy as np

# Define the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Instantiate the global model (needs the same architecture as the trained model)
input_dim = X_test_tensor.shape[-1]
global_model = GRUAutoencoder(input_dim=input_dim).to(device)

# 1. Load the saved global model state dictionary
global_model.load_state_dict(torch.load("gru_ae_fed_global_torch.pth", map_location=device))

# 2. Set the model to evaluation mode
global_model.eval()

# 3. Calculate reconstruction errors for the entire test dataset
criterion = nn.MSELoss(reduction='none') # Use reduction='none' to get per-sample loss
per_sample_mse_test = []

with torch.no_grad():
    for batch in test_loader:
        batch_data = batch[0].to(device)
        output = global_model(batch_data)
        # Calculate MSE for each sample in the batch
        # Output and batch_data are (batch_size, seq_len, input_dim)
        # criterion(output, batch_data) will be (batch_size, seq_len, input_dim)
        # Mean over seq_len and input_dim to get (batch_size,)
        mse = criterion(output, batch_data).mean(dim=(1, 2))
        per_sample_mse_test.extend(mse.cpu().numpy())

# Convert list to numpy array
per_sample_mse_test = np.array(per_sample_mse_test)

# 4. Compute the mean and standard deviation of the reconstruction errors
mean_recon_error = np.mean(per_sample_mse_test)
std_recon_error = np.std(per_sample_mse_test)

# 5. Print the mean and standard deviation
print(f"\nMean Reconstruction Error on Test Set: {mean_recon_error:.6f}")
print(f"Standard Deviation of Reconstruction Error on Test Set: {std_recon_error:.6f}")
```

## Use reconstruction error for anomaly detection

### Subtask:
Use the reconstruction error from the trained autoencoder to refine anomaly detection.


**Reasoning**:
Calculate the reconstruction errors for the training data, determine the anomaly threshold based on training errors, and classify the test data using this threshold.



```python
# Calculate reconstruction errors for the training data
per_sample_mse_train = []
global_model.eval() # Ensure model is in evaluation mode
with torch.no_grad():
    for batch in train_loader:
        # train_loader yields both features and labels, we only need features
        batch_data = batch[0].to(device)
        output = global_model(batch_data)
        mse = criterion(output, batch_data).mean(dim=(1, 2))
        per_sample_mse_train.extend(mse.cpu().numpy())

per_sample_mse_train = np.array(per_sample_mse_train)

# Determine anomaly threshold based on training reconstruction errors
# Using mean + 3 * std
mean_train_error = np.mean(per_sample_mse_train)
std_train_error = np.std(per_sample_mse_train)
anomaly_threshold_ae = mean_train_error + 3 * std_train_error

# Classify test data using the anomaly threshold
y_pred_ae = (per_sample_mse_test >= anomaly_threshold_ae).astype(int)

# Print the calculated anomaly threshold
print(f"\nAnomaly threshold based on training reconstruction error (Mean + 3*Std): {anomaly_threshold_ae:.6f}")
```

```python
# Full fixed script: GRU Autoencoder -> per-feature errors -> RF (AE+RF) vs RF on raw features -> SHAP explainability + metrics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (
    confusion_matrix, accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, ConfusionMatrixDisplay
)
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import shap
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, GRU, RepeatVector, TimeDistributed, Dense
import warnings
warnings.filterwarnings("ignore")
np.random.seed(42)
tf.random.set_seed(42)

# === Load data ===
df = pd.read_csv("merged_user_features.csv")

# Drop obviously non-numeric columns (adjust names if different)
df = df.drop(columns=["LastLoginTime"], errors="ignore")

# Encode object columns
for col in df.select_dtypes(include="object").columns:
    df[col] = LabelEncoder().fit_transform(df[col].astype(str))

# Labels: use existing 'label' or create synthetic for demonstration
if "label" in df.columns:
    y = df["label"].values
    X = df.drop(columns=["label"])
else:
    # synthetic labels (imbalanced example can be created if desired)
    y = np.random.randint(0, 2, len(df))
    X = df.copy()

feature_names = X.columns.tolist()
num_features = X.shape[1]

# === Scale features ===
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # shape (N, F)

# === Prepare sequences for GRU ===
# We'll treat each sample as a sequence of length 1 with 'num_features' features:
# shape -> (N, timesteps=1, features=num_features)
X_seq = X_scaled.reshape((-1, 1, num_features))

# Train/test split (simple contiguous split)
n = len(X_seq)
split = int(0.8 * n)
X_train, X_test = X_seq[:split], X_seq[split:]
y_train, y_test = y[:split], y[split:]

timesteps = X_train.shape[1]  # 1
features = X_train.shape[2]   # num_features

# === Build GRU Autoencoder ===
latent_dim = 32

inp = Input(shape=(timesteps, features))
# Encoder: GRU over the 1-timestep sequence -> produces latent vector of size latent_dim
encoded = GRU(latent_dim, activation="tanh")(inp)
# Decoder: repeat vector and GRU to reconstruct sequence
decoded = RepeatVector(timesteps)(encoded)
decoded = GRU(latent_dim, activation="tanh", return_sequences=True)(decoded)
out = TimeDistributed(Dense(features))(decoded)

autoencoder = Model(inp, out)
autoencoder.compile(optimizer="adam", loss="mse")
autoencoder.summary()

# === Train autoencoder ===
EPOCHS_AE = 10  # increase if you want better reconstructions
BATCH = 32
autoencoder.fit(
    X_train, X_train,
    epochs=EPOCHS_AE,
    batch_size=BATCH,
    validation_split=0.1,
    verbose=1
)

# === Get reconstructions and per-feature squared errors ===
recon_train = autoencoder.predict(X_train)  # shape (N_train, 1, F)
recon_test = autoencoder.predict(X_test)    # shape (N_test, 1, F)

# Per-feature squared error (not averaged yet) -> shape (N, F)
per_feature_err_tr = np.mean(np.square(X_train - recon_train), axis=1)  # average over timesteps (timesteps=1)
per_feature_err_te = np.mean(np.square(X_test - recon_test), axis=1)

# Optionally, you could use raw squared error (without mean) if timesteps>1 you'd average across timesteps.
# For convenience, we now have (N, F) arrays that align with 'feature_names'

# === Train classifiers ===
# 1) RF on reconstruction errors (AE+RF)
rf_ae = RandomForestClassifier(n_estimators=200, random_state=42)
rf_ae.fit(per_feature_err_tr, y_train)
y_pred_ae = rf_ae.predict(per_feature_err_te)
y_proba_ae = rf_ae.predict_proba(per_feature_err_te)[:, 1] if hasattr(rf_ae, "predict_proba") else None

# 2) RF on raw scaled features (baseline RF)
rf_raw = RandomForestClassifier(n_estimators=200, random_state=42)
rf_raw.fit(X_train.reshape(X_train.shape[0], -1), y_train)
y_pred_raw = rf_raw.predict(X_test.reshape(X_test.shape[0], -1))
y_proba_raw = rf_raw.predict_proba(X_test.reshape(X_test.shape[0], -1))[:, 1] if hasattr(rf_raw, "predict_proba") else None

# 3) Logistic Regression on raw features (as another baseline)
lr_raw = LogisticRegression(max_iter=1000, random_state=42)
lr_raw.fit(X_train.reshape(X_train.shape[0], -1), y_train)
y_pred_lr = lr_raw.predict(X_test.reshape(X_test.shape[0], -1))
y_proba_lr = lr_raw.predict_proba(X_test.reshape(X_test.shape[0], -1))[:, 1] if hasattr(lr_raw, "predict_proba") else None

# === Evaluation helper ===
def print_metrics(y_true, y_pred, y_proba=None, name="Model"):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    auc = roc_auc_score(y_true, y_proba) if (y_proba is not None and len(np.unique(y_true))>1) else np.nan
    print(f"\n{name} metrics:")
    print(f"  Accuracy: {acc:.4f}")
    print(f"  Precision: {prec:.4f}")
    print(f"  Recall: {rec:.4f}")
    print(f"  F1: {f1:.4f}")
    if not np.isnan(auc):
        print(f"  ROC AUC: {auc:.4f}")
    cm = confusion_matrix(y_true, y_pred)
    print(f"  Confusion Matrix:\n{cm}")
    return {"acc": acc, "precision": prec, "recall": rec, "f1": f1, "auc": auc, "cm": cm}

metrics_ae = print_metrics(y_test, y_pred_ae, y_proba_ae, name="AE + RF (using reconstruction errors)")
metrics_raw_rf = print_metrics(y_test, y_pred_raw, y_proba_raw, name="RF on raw features")
metrics_lr = print_metrics(y_test, y_pred_lr, y_proba_lr, name="LogisticRegression on raw features")

# Display confusion matrices
plt.figure(figsize=(14,4))
plt.subplot(1,3,1)
ConfusionMatrixDisplay(metrics_ae["cm"]).plot(ax=plt.gca())
plt.title("AE + RF")
plt.subplot(1,3,2)
ConfusionMatrixDisplay(metrics_raw_rf["cm"]).plot(ax=plt.gca())
plt.title("RF raw")
plt.subplot(1,3,3)
ConfusionMatrixDisplay(metrics_lr["cm"]).plot(ax=plt.gca())
plt.title("LR raw")
plt.tight_layout()
plt.show()

# === SHAP explainability ===
# We'll produce SHAP for:
#  - rf_ae (features = reconstruction-error-per-original-feature)
#  - rf_raw (features = original features)
#
# We will show:
#  - Global summary plot (which features overall matter)
#  - Single-sample bar (contribution to prediction) for one anomaly sample if available

# Helper to compute and plot SHAP for a tree model
def shap_for_tree(model, background_data, test_data, feature_names, model_label="Model"):
    # Background selection (max 200)
    bg = background_data
    if bg.shape[0] > 200:
        idx = np.random.choice(bg.shape[0], 200, replace=False)
        bg = bg[idx]
    explainer = shap.TreeExplainer(model, data=bg)
    # Subset test for speed (up to 200)
    subset_n = min(200, test_data.shape[0])
    idxs = np.random.choice(test_data.shape[0], subset_n, replace=False)
    subset = test_data[idxs]
    shap_vals = explainer.shap_values(subset)
    # For binary classifiers, shap_vals is list-like (class0, class1) -> we want positive class (1)
    if isinstance(shap_vals, list) or isinstance(shap_vals, tuple):
        if len(shap_vals) > 1:
            shap_to_plot = shap_vals[1]
        else:
            shap_to_plot = shap_vals[0]
    else:
        shap_to_plot = shap_vals

    # Global summary plot
    print(f"\nSHAP summary for {model_label} (showing top features)...")
    try:
        shap.summary_plot(shap_to_plot, subset, feature_names=feature_names, show=True)
    except Exception as e:
        print("Error plotting SHAP summary:", e)

    # Single-sample: pick an anomaly if possible, else first sample
    return explainer, shap_to_plot, idxs

# SHAP for AE+RF (feature names are original features but represent reconstruction errors)
explainer_ae, shap_to_plot_ae, subset_idxs_ae = shap_for_tree(
    rf_ae, per_feature_err_tr, per_feature_err_te, feature_names, model_label="AE + RF"
)

# SHAP for RF on raw features
explainer_raw, shap_to_plot_raw, subset_idxs_raw = shap_for_tree(
    rf_raw, X_train.reshape(X_train.shape[0], -1), X_test.reshape(X_test.shape[0], -1), feature_names, model_label="RF raw"
)

# === Single-sample SHAP bar for AE+RF and RF raw ===
# Find one anomalous test sample index if exists
anomaly_inds = np.where(y_test == 1)[0]
sample_local_idx = int(anomaly_inds[0]) if anomaly_inds.size > 0 else 0

# For AE+RF: compute SHAP for that sample (needs same shape as training data for explainer)
sample_ae = per_feature_err_te[sample_local_idx].reshape(1, -1)
shap_single_ae = explainer_ae.shap_values(sample_ae)
if isinstance(shap_single_ae, list) and len(shap_single_ae) > 1:
    shap_single_ae = shap_single_ae[1].flatten()
elif isinstance(shap_single_ae, list):
    shap_single_ae = shap_single_ae[0].flatten()
else:
    shap_single_ae = np.array(shap_single_ae).flatten()

# For RF raw:
sample_raw = X_test.reshape(X_test.shape[0], -1)[sample_local_idx].reshape(1, -1)
shap_single_raw = explainer_raw.shap_values(sample_raw)
if isinstance(shap_single_raw, list) and len(shap_single_raw) > 1:
    shap_single_raw = shap_single_raw[1].flatten()
elif isinstance(shap_single_raw, list):
    shap_single_raw = shap_single_raw[0].flatten()
else:
    shap_single_raw = np.array(shap_single_raw).flatten()

TOP_K = min(12, len(feature_names))

# Plot single-sample bar AE+RF
order_ae = np.argsort(np.abs(shap_single_ae))[::-1][:TOP_K]
plt.figure(figsize=(8,4))
plt.barh([feature_names[i] for i in order_ae[::-1]], shap_single_ae[order_ae][::-1])
plt.title(f"Single-sample SHAP (AE+RF) — sample idx {sample_local_idx}")
plt.xlabel("SHAP value (contribution to prediction)")
plt.tight_layout()
plt.show()

# Plot single-sample bar RF raw
order_raw = np.argsort(np.abs(shap_single_raw))[::-1][:TOP_K]
plt.figure(figsize=(8,4))
plt.barh([feature_names[i] for i in order_raw[::-1]], shap_single_raw[order_raw][::-1])
plt.title(f"Single-sample SHAP (RF raw) — sample idx {sample_local_idx}")
plt.xlabel("SHAP value (contribution to prediction)")
plt.tight_layout()
plt.show()

# --- Static bar plot of mean(|SHAP|) for AE+RF (global importance) ---
mean_abs_shap_ae = np.mean(np.abs(shap_to_plot_ae), axis=0)
idxs = np.argsort(mean_abs_shap_ae)[::-1][:TOP_K]
plt.figure(figsize=(8,4))
plt.barh([feature_names[i] for i in idxs[::-1]], mean_abs_shap_ae[idxs][::-1])
plt.title(f"Top {TOP_K} features by mean |SHAP| (AE+RF)")
plt.xlabel("mean(|SHAP value|)")
plt.tight_layout()
plt.show()

# --- Static bar plot of mean(|SHAP|) for RF raw (global importance) ---
mean_abs_shap_raw = np.mean(np.abs(shap_to_plot_raw), axis=0)
idxs = np.argsort(mean_abs_shap_raw)[::-1][:TOP_K]
plt.figure(figsize=(8,4))
plt.barh([feature_names[i] for i in idxs[::-1]], mean_abs_shap_raw[idxs][::-1])
plt.title(f"Top {TOP_K} features by mean |SHAP| (RF raw)")
plt.xlabel("mean(|SHAP value|)")
plt.tight_layout()
plt.show()

# === Quick model comparison (by F1 and ROC AUC) ===
comparison = pd.DataFrame({
    "model": ["AE+RF", "RF_raw", "LR_raw"],
    "accuracy": [metrics_ae["acc"], metrics_raw_rf["acc"], metrics_lr["acc"]],
    "precision": [metrics_ae["precision"], metrics_raw_rf["precision"], metrics_lr["precision"]],
    "recall": [metrics_ae["recall"], metrics_raw_rf["recall"], metrics_lr["recall"]],
    "f1": [metrics_ae["f1"], metrics_raw_rf["f1"], metrics_lr["f1"]],
    "auc": [metrics_ae["auc"], metrics_raw_rf["auc"], metrics_lr["auc"]],
})
print("\nModel comparison:\n", comparison.sort_values("f1", ascending=False).reset_index(drop=True))

# Optional: save models, shap objects, or export plots as needed.
```
